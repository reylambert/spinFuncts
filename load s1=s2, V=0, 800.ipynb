```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# iterate over s1=s2=50, E=1, V=800, L=[10,0], E=1
# s2, z, phi, Jx1, Jy1, times
d1=np.load("/rhome/mlamb005/spincode/results/equalS_V0.npy", allow_pickle=True)
d2=np.load("/rhome/mlamb005/spincode/results/equalS_V800.npy", allow_pickle=True)
```

```python
z_t_0   = d1[0]
phi_t_0 = d1[1]
Jx_1_0  = d1[2]
Jy_1_0  = d1[3]
times   = d1[4]
#print(np.shape(d1))
#print(len(z_t))
z_t_V   = d2[0]
phi_t_V = d2[1]
Jx_1_V  = d2[2]
Jy_1_V  = d2[3]
```

```python
for i in range(len(z_t_0)):
    plt.plot(times, z_t_0[i], label=phi_t_0[i][0])
plt.ylabel('z_t (V=0)')
plt.xlabel('time')
plt.legend()
```

```python
for i in range(len(z_t_V)):
    plt.plot(times, z_t_V[i], label=phi_t_V[i][0])
plt.ylabel('z_t (V=800)')
plt.xlabel('time')
plt.legend()
```

```python
for i in range(len(z_t_0)):
    plt.plot(times, Jy_1_0[i], label=phi_t_0[i][0])
plt.ylabel('Jy_1, V=0')
plt.xlabel('time')
plt.legend()
```

```python
for i in range(len(z_t_V)):
    plt.plot(times, Jy_1_V[i], label=phi_t_V[i][0])
plt.ylabel('Jy_1, V=800')
plt.xlabel('time')
plt.legend()
```

```python
# take time averages of z, etc and plot those
z_avg_0 = []
for i in range(len(z_t_0)):
    z_avg_0.append(np.average([z_t_0[i][100:]]))
print(z_avg_0)

z_avg_V = []
for i in range(len(z_t_V)):
    z_avg_V.append(np.average([z_t_V[i][100:]]))
print(z_avg_V)
```

```python
plt.plot(z_avg_0, z_avg_V)
plt.ylabel('<z(t)>, V=800')
plt.xlabel('<z(t)>, V=0')
```
