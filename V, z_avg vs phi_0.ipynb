```python
import numpy as np
from matplotlib import pyplot as plt

import spinFuncts as sf
```

```python
E = 1

N  = 1        #system size
V  = 800        #coupling strength
L  = [10, 0]   #lambda prefix, Jzz_i

S1 = '100'    #spin of 1
S2 = ['1/2','2','5','10']    #spin of 2

s1 = 100
s2 = [1/2, 2, 5, 10]
```

```python
tf = s1

phis = [0, np.pi/8, np.pi/4, 3*np.pi/8, np.pi/2, 5*np.pi/8, 3*np.pi/4, 7*np.pi/8, np.pi]
zs   = sf.getZs_Ec(phis, L[0], E)
times = np.linspace(0.0, tf, 1000)

print(zs)
```

```python
z_t   = []
phi_t = []

Jx_1 = []
Jy_1 = []
Jz_1 = []

for i in range(len(phis)):
    zi, phii, J1_i, J2_i = sf.evolveSys_all(zs[i], phis[i], 0, 0, times, S1, s1, S2[1], s2[1], L, V, N)
    z_t.append(zi[0])
    phi_t.append(phii[0])
    Jx_1.append(J1_i[0])
    Jy_1.append(J1_i[1])
    Jz_1.append(J1_i[2])
```

```python
for i in range(len(phis)):
    plt.plot(phi_t[i], z_t[i], label=phis[i])

plt.ylabel('z')
plt.xlabel('phi')
plt.legend()
print("E =", E, ", |J1|=", S1, ", |J2|=", S2[0])
```

```python
for i in range(len(phis)):
    plt.plot(times, z_t[i], label=phis[i])

plt.ylabel('z')
plt.xlabel('time')
plt.legend()
print("E =", E, ", |J1|=", S1, ", |J2|=", S2[0])
```

```python
steps = 1000
tf    = 100
times = np.linspace(0.0, tf, steps)

def slice_avg(i, f, vect, t_t=tf):
    dt = tf/steps
    a  = int(i/dt)
    b  = int(f/dt)
    return sum(vect[a:b])/(f-i)

print(slice_avg(0,100, times))
print(sum(times)/100)
```
