```python
from quspin.operators import hamiltonian
from quspin.basis import spin_basis_general, tensor_basis
from quspin.tools.measurements import obs_vs_time
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize

import spinFuncts as sf
```

```python
E = 1

N  = 1        #system size
V  = 50        #coupling strength
L  = [10, 0]   #lambda prefix, Jzz_i

S1 = '100'    #spin of 1
S2 = '2'    #spin of 2

s1 = 100
s2 = 2
```

```python
tf = s1

phis = [0, np.pi/8, np.pi/4, 3*np.pi/8, np.pi/2, 5*np.pi/8, 3*np.pi/4, 7*np.pi/8, np.pi]
zs   = sf.getZs_Ec(phis, L[0], E)
times = np.linspace(0.0, tf, 1000)

print(zs)
```

```python
# only looking at s1 behavior
z_t   = []
phi_t = []

Jx_1 = []
Jy_1 = []
Jz_1 = []

for i in range(len(phis)):
    zi, phii, J1_i, J2_i = sf.evolveSys_all(zs[i], phis[i], 0, 0, times, S1, s1, S2, s2, L, V, N)
    z_t.append(zi[0])
    phi_t.append(phii[0])
    Jx_1.append(J1_i[0])
    Jy_1.append(J1_i[1])
    Jz_1.append(J1_i[2])
```

```python
for i in range(len(phis)):
    plt.plot(phi_t[i], z_t[i], label=i)

plt.ylabel('z')
plt.xlabel('phi')
plt.legend()
print("E =", E, ", L =", L)
```

```python
for i in range(len(phis)):
    plt.plot(times, z_t[i], label=i)

plt.ylabel('z')
plt.xlabel('time')
plt.legend()
print("E =", E, ", L =", L)
```

```python
for i in range(len(phis)):
    plt.plot(times, Jz_1[i], label=i)

plt.ylabel('Jz_1')
plt.xlabel('time')
plt.legend()
print("E =", E, ", L =", L)
```

```python
for i in range(len(phis)):
    plt.plot(times, Jx_1[i], label=i)

plt.ylabel('Jx_1')
plt.xlabel('time')
plt.legend()
print("E =", E, ", L =", L)
```
