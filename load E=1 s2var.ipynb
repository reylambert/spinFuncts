```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# iterate over s2, E=1, V=800, L=[10,0], E=1, s1=100
# s2, z, phi, Jx1, Jy1, times
d1=np.load("/rhome/mlamb005/spincode/results/V_s2_zpxyt.npy", allow_pickle=True)
```

```python
s2    = d1[0]
z_t   = d1[1]
phi_t = d1[2]
Jx_1  = d1[3]
Jy_1  = d1[4]
times = d1[5]
#print(np.shape(d1))
#print(len(z_t))
```

```python
# plot stuff, for loops
for i in range(len(z_t)):
    plt.plot(times, z_t[i], label=s2[i])
plt.ylabel('z_t')
plt.xlabel('time')
plt.legend()
```

```python
for i in range(len(z_t)):
    plt.plot(times, Jy_1[i], label=s2[i])
plt.ylabel('Jy_1')
plt.xlabel('time')
plt.legend()
```

```python
# take time averages of z, etc and plot those
z_avg = []
for i in range(len(z_t)):
    z_avg.append(np.average([z_t[i][100:]]))
print(z_avg)
```

```python
plt.plot(z_avg, s2)
plt.ylabel('s2, |J2|')
plt.xlabel('<z(t)>')
```
